# Step2. Count the total number of freegoups of a braid

```python
# vectorize the eval function
ev = np.vectorize(eval)

# count free_part of the dataset
ff = np.vectorize(count_FP)
df_10000_9['total_num_FP'] = ff(ev(df_10000_9['free_part']))

# count torsion_part of the dataset
ft = np.vectorize(count_TP)
df_10000_9['total_num_TP'] = ft(ev(df_10000_9['torsion_part']))
```

# Step3. Count the total number of freegroups for a braid per row

```python
# find maximum and minimum row number
find_max_row = float('-inf')
find_min_row = float('inf')
for index, a_row in df_10000_9.iterrows():
    #print(a_row.free_part)
    for a_key in eval(a_row.free_part):
       #print(a_key[0])
        if a_key[0] < find_min_row:
            find_min_row = a_key[0]
        if a_key[0] > find_max_row:
            find_max_row = a_key[0]

find_max_row, find_min_row
```
